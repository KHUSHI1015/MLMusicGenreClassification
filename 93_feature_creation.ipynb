```python
zcr = {'mean' : [],
'median' : [],
'std' : []
}
s_c = {'mean' : [],
'median' : [],
'std' : []
}
s_co = {'mean' : [],
'median' : [],
'std' : []
}
s_b = {'mean' : [],
'median' : [],
'std' : []
}
s_ro = {'mean' : [],
'median' : [],
'std' : []
}
mfcc = [{'mean' : [],
'median' : [],
'std' : []
} for i in range(13)]
mfcc_d = [{'mean' : [],
'median' : [],
'std' : []
}for i in range(13)]
```

```python
Y = []

base_path = 'Data\\genres_original'
count = 0
#looping through all label directories
for label in tqdm(os.listdir(base_path)):
    file_path = base_path + '/' + label
    
    #looping through each file in the directory
    for pth in os.listdir(file_path):
        final_path = file_path + '/' + pth
        try:
        #loading original file
            audio, sr = librosa.load(final_path,duration = 28)

            s1 = librosa.feature.zero_crossing_rate(audio)[0]
            s2 = librosa.feature.spectral_centroid(audio)[0]
            s3 = librosa.feature.spectral_contrast(audio)[0]
            s4 = librosa.feature.spectral_bandwidth(audio)[0]
            s5 = librosa.feature.spectral_rolloff(audio)[0]
            s6 = librosa.feature.mfcc(audio)
            s7 = librosa.feature.delta(s6)

            zcr['mean'].append(statistics.mean(s1)) 
            zcr['median'].append(statistics.median(s1)) 
            zcr['std'].append(statistics.stdev(s1))

            s_c['mean'].append(statistics.mean(s2))
            s_c['median'].append(statistics.median(s2))
            s_c['std'].append(statistics.stdev(s2))

            s_co['mean'].append(statistics.mean(s3))
            s_co['median'].append(statistics.median(s3))
            s_co['std'].append(statistics.stdev(s3))

            s_b['mean'].append(statistics.mean(s4))
            s_b['median'].append(statistics.median(s4))
            s_b['std'].append(statistics.stdev(s4))

            s_ro['mean'].append(statistics.mean(s5))
            s_ro['median'].append(statistics.median(s5))
            s_ro['std'].append(statistics.stdev(s5))

            for i in range(13):
                mfcc[i]['mean'].append(statistics.mean(s6[i]))
                mfcc[i]['median'].append(statistics.median(s6[i]))
                mfcc[i]['std'].append(statistics.stdev(s6[i]))

                mfcc_d[i]['mean'].append(statistics.mean(s7[i]))
                mfcc_d[i]['median'].append(statistics.median(s7[i]))
                mfcc_d[i]['std'].append(statistics.stdev(s7[i]))

            #appending labels to the label list
            Y.append(label)
            print(count, label)
            count += 1
        except:
            print("Error in:",final_path)
```
